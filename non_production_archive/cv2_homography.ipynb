```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
#%matplotlib inline  # if you are running this code in Jupyter notebook
```

```python
img_captured = cv2.imread('empy_chessboard_top_view.jpg')
img_captured = cv2.imread('test_chessboard.jpg')

plt.imshow(img_captured)
```

```python
lwr = np.array([0, 0, 143])
upr = np.array([179, 61, 252])
hsv = cv2.cvtColor(img_captured, cv2.COLOR_BGR2HSV)
msk = cv2.inRange(hsv, lwr, upr)

plt.imshow(msk)
```

```python
krn = cv2.getStructuringElement(cv2.MORPH_RECT, (50, 30))
dlt = cv2.dilate(msk, krn, iterations=5)
res = 255 - cv2.bitwise_and(dlt, msk)

plt.imshow(res)
```

```python
#img_captured = cv2.imread('test_chessboard.jpg')
#img_captured = cv2.imread('empy_chessboard_top_view.jpg')
img_captured = cv2.imread('test.jpg')
plt.imshow(img_captured)
res = np.uint8(res)

pattern = (7, 7)
ret, corners = cv2.findChessboardCorners(img_captured, pattern,
                                         flags=cv2.CALIB_CB_ADAPTIVE_THRESH +
                                               cv2.CALIB_CB_FAST_CHECK +
                                               cv2.CALIB_CB_NORMALIZE_IMAGE)

#ret, corners = cv2.findChessboardCornersSB(img_captured, pattern, flags=cv2.CALIB_CB_EXHAUSTIVE)
if ret:
    #print(corners)
    fnl = cv2.drawChessboardCorners(img_captured, pattern, corners, ret)
    plt.imshow(fnl)
    cv2.waitKey(0)
else:
    print("No Checkerboard Found")
```
